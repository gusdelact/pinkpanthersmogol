## 3. Configuración del modelo LLM

```python
from smolagents import Tool, ToolCallingAgent, LiteLLMModel

model = LiteLLMModel(
    model_id="bedrock/converse/us.anthropic.claude-sonnet-4-20250514-v1:0",
    aws_region_name="us-east-1",
)
```

## 4. Definición de la tripulación de agentes (Tools)

Cada tool representa un personaje de El Principito con una especialidad diferente.

```python
class ReyEstrategaTool(Tool):
    name = "rey_estratega"
    description = "Agente experto en estrategia de negocios para aconsejar a Pink Panther."
    inputs = {
        "situacion": {
            "type": "string",
            "description": "Situación o problema narrativo que enfrenta Pink Panther."
        }
    }
    output_type = "string"

    def forward(self, situacion: str) -> str:
        return (
            "El Rey aconseja pensar en propósito, ventaja estratégica y alianzas. "
            f"Para esta situación: {situacion}, Pink Panther debe definir primero "
            "qué reino quiere proteger y qué costo está dispuesta a pagar."
        )
```

```python
class VanidosoArteTool(Tool):
    name = "vanidoso_arte"
    description = "Agente experto en arte, estética y narrativa visual."
    inputs = {
        "escena": {
            "type": "string",
            "description": "Escena que Pink Panther quiere embellecer o interpretar."
        }
    }
    output_type = "string"

    def forward(self, escena: str) -> str:
        return (
            "El Vanidoso observa la escena como un cuadro: composición, color, símbolo y gesto. "
            f"Para la escena: {escena}, recomienda usar contraste entre elegancia rosa "
            "y sombras azules para crear misterio."
        )
```

```python
class HombreNegociosQuantTool(Tool):
    name = "hombre_negocios_quant"
    description = "Agente experto en finanzas cuantitativas y riesgo."
    inputs = {
        "problema": {
            "type": "string",
            "description": "Problema financiero, de riesgo o decisión bajo incertidumbre."
        }
    }
    output_type = "string"

    def forward(self, problema: str) -> str:
        return (
            "El Hombre de Negocios propone medir escenarios, volatilidad y pérdida esperada. "
            f"Para el problema: {problema}, Pink Panther debe simular varios futuros "
            "antes de decidir."
        )
```

```python
class FaroleroEnergiaTool(Tool):
    name = "farolero_energia"
    description = "Agente experto en energía eléctrica, continuidad operativa y resiliencia."
    inputs = {
        "incidente": {
            "type": "string",
            "description": "Incidente energético o de infraestructura."
        }
    }
    output_type = "string"

    def forward(self, incidente: str) -> str:
        return (
            "El Farolero recomienda redundancia, monitoreo y encendido disciplinado de cada farol. "
            f"Ante el incidente: {incidente}, Pink Panther debe priorizar continuidad, "
            "seguridad y recuperación gradual."
        )
```

```python
class GeografoFantasiaTool(Tool):
    name = "geografo_fantasia"
    description = "Agente experto en mundos de fantasía, mapas, territorios y lore."
    inputs = {
        "mundo": {
            "type": "string",
            "description": "Descripción del territorio fantástico."
        }
    }
    output_type = "string"

    def forward(self, mundo: str) -> str:
        return (
            "El Geógrafo dibuja montañas, ríos, ciudades antiguas y fronteras invisibles. "
            f"Para el mundo: {mundo}, sugiere crear un mapa con cinco regiones: "
            "el Reino, la Galería, el Mercado, la Torre del Farol y el Bosque Rosa."
        )
```

## 5. Creación del agente orquestador

```python
pink_panther_agent = ToolCallingAgent(
    tools=[
        ReyEstrategaTool(),
        VanidosoArteTool(),
        HombreNegociosQuantTool(),
        FaroleroEnergiaTool(),
        GeografoFantasiaTool(),
    ],
    model=model,
    max_steps=6,
)
```

## 6. Ejecución — La aventura de Pink Panther

```python
respuesta = pink_panther_agent.run(
    """
    Pink Panther llega a un mundo de fantasía donde la ciudad perdió energía,
    el mercado financiero del reino está en pánico y una galería de arte guarda
    un mapa secreto. Pide consejo a los agentes especializados y construye
    un plan de aventura.
    """
)
```
